```python
import time
import matplotlib.pyplot as plt
# Iterative algorithm
def iterative_sum(N):
    result = 0
    for i in range(1, N+1):
        result += i
    return result
# Recursive algorithm
def recursive_sum(N):
    if N == 0:
        return 0
    else:
        return N + recursive_sum(N-1)
# Function to measure execution time
def measure_time(func, N):
    start_time = time.time()
    func(N)
    end_time = time.time()
    return end_time - start_time
# List to store execution times
iterative_times = []
recursive_times = []
# Values of N
N_values = list(range(1, 1001, 100))  
# Measure execution times
for N in N_values:
    iterative_times.append(measure_time(iterative_sum, N))
    recursive_times.append(measure_time(recursive_sum, N))
plt.plot(N_values, iterative_times, label='Iterative')
plt.plot(N_values, recursive_times, label='Recursive')
plt.xlabel('Value of N')
plt.ylabel('Execution Time (seconds)')
plt.title('Execution Time of Iterative vs Recursive Sum Calculation')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import time
import random
import matplotlib.pyplot as plt
# Function to generate a random array
def generate_random_array(size):
    return [random.randint(1, 1000) for _ in range(size)]
# Linear search algorithm
def linear_search(arr, key):
    for i in range(len(arr)):
        if arr[i] == key:
            return i
    return -1
# Binary search algorithm
def binary_search(arr, key):
    low = 0
    high = len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == key:
            return mid
        elif arr[mid] < key:
            low = mid + 1
        else:
            high = mid - 1
    return -1
# Function to measure execution time
def measure_time(func, arr, key):
    start_time = time.time()
    func(arr, key)
    end_time = time.time()
    return end_time - start_time
# List to store execution times
linear_times = []
binary_times = []
arr = generate_random_array(10000)
search_keys = [random.randint(1, 1000) for _ in range(5)]
for key in search_keys:
    linear_times.append(measure_time(linear_search, arr, key))
    binary_times.append(measure_time(binary_search, sorted(arr), key))
plt.plot(range(1, 6), linear_times, label='Linear Search')
plt.plot(range(1, 6), binary_times, label='Binary Search')
plt.xlabel('Search Key')
plt.ylabel('Execution Time (seconds)')
plt.title('Execution Time of Linear vs Binary Search')
plt.xticks(range(1, 6))
plt.legend()
plt.grid(True)
plt.show()
```

```python
def string_to_integer(s):
    if not s:
        return 0
    digit = int(s[0])
    # Recursive call to convert the rest of the string
    remaining_digits = string_to_integer(s[1:])
    # Multiply the integer value of the first digit with 10 raised to the power of the remaining string length
    return digit * (10 ** len(s[1:])) + remaining_digits
# Testing the function
input_string = "13531"
result = string_to_integer(input_string)
print(f"The integer representation of '{input_string}' is:", result)
```

```python
def reverse_string(s):
    if len(s) <= 1:
        return s
    # Recursive case: reverse the substring starting from the second character and concatenate the first character
    return reverse_string(s[1:]) + s[0]
# Testing the function
input_string = "pots&pans"
reversed_string = reverse_string(input_string)
print(f"The reverse of '{input_string}' is:", reversed_string)
```

```python
def is_palindrome(s):
    if len(s) <= 1:
        return True
    if s[0] == s[-1]:
        # Recursive call: check if the substring excluding the first and last characters is a palindrome
        return is_palindrome(s[1:-1])
    else:
        # If the first and last characters are not equal, it's not a palindrome
        return False

# Test the function
palindrome_examples = ["racecar", "gohangasalamiimalasagnahog", "hello", "level", "radar"]
for example in palindrome_examples:
    print(f"'{example}' is a palindrome:", is_palindrome(example))
```
